```python
import csv
import statistics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def getTimes(filename):
    timesList = []
    
    with open(filename) as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            timesList.append(float(row[0]))
    return timesList

rls=getTimes("release_time")
dbb=getTimes("debug_time")

print(f"Release Mean= {statistics.mean(rls)}, stddev={statistics.stdev(rls)}")
print(f"Debug Mean= {statistics.mean(dbb)}, stddev={statistics.stdev(dbb)}")
```

```python
plt.plot(dbb)
plt.title('Tempo de execução da simulação, BUILD_TYPE=Debug')
plt.xlabel('Número da execução')
plt.ylabel('Segundos')
```

```python
plt.hist(dbb, bins=20)
plt.title('Tempo de execução da simulação, BUILD_TYPE=Debug')
plt.xlabel('Duração (segundos)')
plt.ylabel('Frequência')
```

```python
plt.hist(rls, bins=20)
plt.title('Tempo de execução da simulação, BUILD_TYPE=Release')
plt.xlabel('Duração (segundos)')
plt.ylabel('Frequência')
```
